```python
import numpy as np
from PIL import Image

def create_grayscale_3d(image_path):
    # Cargar la imagen y convertirla a escala de grises
    gray_image = Image.open(image_path).convert('L')
    
    # Obtener la matriz de la imagen en escala de grises
    gray_array = np.array(gray_image)
    
    # Repetir la matriz en tres dimensiones para cada canal de color (rojo, verde y azul)
    red_image = np.repeat(gray_array[:, :, np.newaxis], 3, axis=2)
    green_image = np.repeat(gray_array[:, :, np.newaxis], 3, axis=2)
    blue_image = np.repeat(gray_array[:, :, np.newaxis], 3, axis=2)
    
    # Crear imágenes en escala de grises en tres dimensiones a partir de las matrices NumPy
    red_image = Image.fromarray(np.uint8(red_image))
    green_image = Image.fromarray(np.uint8(green_image))
    blue_image = Image.fromarray(np.uint8(blue_image))
    
    return red_image, green_image, blue_image

# Ruta de la imagen
image_path = "C:/Users/Usuario/Documents/U GALILEO/MBI&A/POSGRADO PREDICCION CIENCIA DE DATOS/3r trimestre/1_txt mining imagen/LAB 1/lion.jpg"

# Crear escala de grises en tres dimensiones
red_image, green_image, blue_image = create_grayscale_3d(image_path)

# Mostrar las imágenes generadas
red_image.show()
green_image.show()
blue_image.show()
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

def calculate_grayscale(image):
    # Convertir la imagen a escala de grises mediante el promedio aritmético
    red_band, green_band, blue_band = image.split()
    gray_band = (np.array(red_band) + np.array(green_band) + np.array(blue_band)) // 3
    return Image.fromarray(gray_band)

def weighted_mean(data, weights):
    return np.sum(data * weights) / np.sum(weights)

def show_histograms(image_path):
    # Cargar la imagen
    image = Image.open(image_path)
    
    # Convertir la imagen a escala de grises
    gray_image = calculate_grayscale(image)
    
    # Obtener los histogramas de cada canal de color y el histograma de escala de grises
    red_hist = image.histogram()[0:256]
    green_hist = image.histogram()[256:512]
    blue_hist = image.histogram()[512:768]
    gray_hist = gray_image.histogram()
    
    # Calcular la media de la distribución
    red_mean = weighted_mean(np.arange(256), red_hist)
    green_mean = weighted_mean(np.arange(256), green_hist)
    blue_mean = weighted_mean(np.arange(256), blue_hist)
    gray_mean = weighted_mean(np.arange(256), gray_hist)
    
    # Mostrar los histogramas con las líneas verticales que indican la media de la distribución
    plt.figure(figsize=(12, 6))
    
    plt.subplot(2, 2, 1)
    plt.plot(red_hist, color='red')
    plt.axvline(red_mean, color='red', linestyle='dashed', linewidth=1)
    plt.title("Histograma Canal Rojo")
    
    plt.subplot(2, 2, 2)
    plt.plot(green_hist, color='green')
    plt.axvline(green_mean, color='green', linestyle='dashed', linewidth=1)
    plt.title("Histograma Canal Verde")
    
    plt.subplot(2, 2, 3)
    plt.plot(blue_hist, color='blue')
    plt.axvline(blue_mean, color='blue', linestyle='dashed', linewidth=1)
    plt.title("Histograma Canal Azul")
    
    plt.subplot(2, 2, 4)
    plt.plot(gray_hist, color='black')
    plt.axvline(gray_mean, color='black', linestyle='dashed', linewidth=1)
    plt.title("Histograma Escala de Grises")
    
    plt.tight_layout()
    plt.show()

# Ejemplo de uso:
image_path = "C:/Users/Usuario/Documents/U GALILEO/MBI&A/POSGRADO PREDICCION CIENCIA DE DATOS/3r trimestre/1_txt mining imagen/LAB 1/italy.jpg"
show_histograms(image_path)

```
