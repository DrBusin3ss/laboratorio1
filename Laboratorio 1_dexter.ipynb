```python
from PIL import Image

def filter_color(image_path, color):
    # Cargamos la imagen
    image = Image.open(image_path)

    # Convertimos la imagen en modo RGB para poder acceder a los canales de color
    image = image.convert('RGB')
    
    # Creamos una imagen en blanco del mismo tamaño para almacenar los resultados
    result_image = Image.new('RGB', image.size)

    # Obtenemos los tamaños de la imagen
    width, height = image.size

    # Iteramos sobre cada pixel de la imagen
    for x in range(width):
        for y in range(height):
            # Obtenemos el valor RGB del pixel actual
            r, g, b = image.getpixel((x, y))

            # Dependiendo del valor del parámetro 'color', activamos o desactivamos los canales
            if color == 1:  # Mostrar solo el color azul
                r, g = 0, 0
            elif color == 2:  # Mostrará solo el color verde
                r, b = 0, 0
            elif color == 3:  # Mostrará solo el color rojo
                g, b = 0, 0
            elif color == 10:  # Mostrará solo rojo y verde
                b = 0
            elif color == 20:  # Mostrará solo verde y azul
                r = 0
            elif color == 30:  # Mostrará solo azul y rojo
                g = 0

            # Establecemos el valor RGB modificado en la imagen de resultado
            result_image.putpixel((x, y), (r, g, b))

    return result_image


```

```python
# 3 para ROJO
filtered_image = filter_color('dexter.jpg', 3)
filtered_image.show()
```

```python
# 1 para AZUL
filtered_image = filter_color('dexter.jpg', 1)
filtered_image.show()
```
